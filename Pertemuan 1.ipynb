### Check Value

```python
#cek jumlah data NaN
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
#mencari rata2
mean_age = df['Age'].mean()
mean_age
```

### Mengisi nilai NaN

```python
#mengisi nilai NaN numerik dengan mean
df['Age'] = df['Age'].fillna(mean_age)
df.isna().sum()
```

```python
#Mencari mode (nilai yang sering muncul)
mode_embarked = df['Embarked'].mode()
mode_embarked
```

```python
#mengisi nilai NaN kategori dengan mode (nilai yang paling banyak keluar)
#[0] adalah pilihan mode yang pertama
df['Embarked'] = df['Embarked'].mode()[0]
df.isna().sum()
```

### Mengganti tipe data

```python
df['AGE'] = df['AGE'].astype('int64')
df.dtypes
```

### Rename columns

```python
new_columns = {
 'PassengerId': 'PASS_ID',
 'Survived': 'SURVIVED', 
 'Pclass': 'PCLASS', 
 'Name': 'NAME', 
 'Sex': 'SEX', 
 'Age': 'AGE', 
 'SibSp': 'SIBSA', 
 'Parch': 'PARCA' , 
 'Fare': 'FARE',
 'Ticket' : 'TICKET',
 'Embarked' : "EMBARKED"
}
```

```python
df = df.rename(columns = new_columns)
df.head()
```

### Mengganti index

```python
df = df.set_index('PASS_ID')
df.head()
```

### Grouping data

```python
df.groupby('PCLASS').sum()
```

```python
df_pclass = df.groupby('PCLASS')['SURVIVED'].sum()
df_pclass
```

```python
#Menampilkan data menggunakan plot
#barh = plot bar horizontal, (X, Y)
plt.barh(df_pclass.index, df_pclass)

#mengubah keterangan sumbu y
plt.yticks(df_pclass.index, ['First Class', 'Second Class', 'Third Class'])

#menampilkan plot
plt.show()
```
